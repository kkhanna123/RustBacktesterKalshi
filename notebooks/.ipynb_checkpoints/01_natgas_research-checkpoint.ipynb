# Kalshi NatGas — Research Notebook

The local analogue of infra-orchestrator's `research-jupyter`. Explores the orderbook deltas the
collector streamed into **ClickHouse**, reconstructs the live book, views a backtest, and computes
binary-option **greeks** — all from Python. Figures are saved to `figures/03_research_notebook/`.

Prereqs: the ClickHouse server is up (`make ch-server`) and the collector has run for a while.

```python
import os, sys, json, datetime as dt
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import requests

ROOT = os.path.abspath('..')
sys.path.insert(0, ROOT)
FIGDIR = os.path.join(ROOT, 'figures', '03_research_notebook')
os.makedirs(FIGDIR, exist_ok=True)
CH = os.environ.get('CLICKHOUSE_URL', 'http://localhost:8123')

def ch(sql):
    """Run a ClickHouse SQL query, return rows as a list of tab-split lists (or [] on error)."""
    try:
        r = requests.post(CH, params={'query': sql + ' FORMAT TabSeparated'}, timeout=15)
        if r.status_code != 200:
            print('CH error:', r.text[:200]); return []
        return [line.split('\t') for line in r.text.strip().splitlines() if line]
    except Exception as e:
        print('ClickHouse unreachable:', e); return []

print('ClickHouse:', ch('SELECT version()'))
```

## 1. What did the collector capture?
List the most-active NatGas instruments in the `orderbook_deltas` table.

```python
rows = ch("SELECT instrument_id, count() c, min(timestamp), max(timestamp) "
          "FROM kalshi.orderbook_deltas WHERE instrument_id LIKE 'KXNATGASD-%' "
          "GROUP BY instrument_id ORDER BY c DESC LIMIT 10")
for inst, c, lo, hi in rows:
    print(f'{inst:32s} {int(c):6d} deltas   {lo[11:19]}..{hi[11:19]}')
top_inst = rows[0][0] if rows else None
print('\nmost active:', top_inst)
```

## 2. Reconstruct and plot the live top-of-book
Replay the deltas (respecting `is_snapshot` resets) into a YES-native book and track best bid/ask/mid.

```python
def reconstruct(inst):
    data = ch("SELECT toUnixTimestamp64Nano(timestamp), action, side, price, size, is_snapshot "
              f"FROM kalshi.orderbook_deltas WHERE instrument_id = '{inst}' "
              "ORDER BY timestamp, sequence")
    book = {'BUY': {}, 'SELL': {}}; out = []
    for ts, action, side, price, size, snap in data:
        price, size = float(price), float(size)
        if int(snap): book = {'BUY': {}, 'SELL': {}}
        if action == 'DELETE' or size <= 0: book[side].pop(price, None)
        else: book[side][price] = size
        bid = max(book['BUY']) if book['BUY'] else None
        ask = min(book['SELL']) if book['SELL'] else None
        mid = (bid+ask)/2 if bid and ask else None
        out.append((int(ts), bid, ask, mid))
    return out

if top_inst:
    tob = reconstruct(top_inst)
    xs = [dt.datetime.utcfromtimestamp(t/1e9) for t,*_ in tob]
    fig, ax = plt.subplots(figsize=(11,4))
    ax.plot(xs, [b for _,b,_,_ in tob], lw=.8, color='#22c55e', label='bid')
    ax.plot(xs, [a for _,_,a,_ in tob], lw=.8, color='#ef4444', label='ask')
    ax.plot(xs, [m for _,_,_,m in tob], lw=1.3, color='#3b82f6', label='mid')
    ax.set_ylim(0,1); ax.set_title(f'Live top-of-book: {top_inst} ({len(tob)} deltas)')
    ax.legend(); ax.grid(alpha=.2); fig.autofmt_xdate()
    fig.savefig(os.path.join(FIGDIR,'fig1_live_book.png'), bbox_inches='tight', dpi=110)
    print('saved fig1_live_book.png')
else:
    print('no collected data yet — start the collector and rerun')
```

## 3. View a backtest result
Load an existing Rust-backtester export and plot the equity curve. (Run a backtest with
`kalshi-backtest backtest ... --out-dir data/exports/<name>` to create one.)

```python
import csv, glob
exp = None
for d in sorted(glob.glob(os.path.join(ROOT,'data','exports','*'))):
    if os.path.exists(os.path.join(d,'equity.csv')) and os.path.exists(os.path.join(d,'report.json')):
        exp = d; break
if exp:
    s = json.load(open(os.path.join(exp,'report.json')))['summary']
    ts, tot = [], []
    with open(os.path.join(exp,'equity.csv')) as f:
        for r in csv.DictReader(f):
            ts.append(dt.datetime.utcfromtimestamp(int(r['ts_ns'])/1e9)); tot.append(float(r['total']))
    fig, ax = plt.subplots(figsize=(11,3.5))
    ax.plot(ts, tot, color='#3b82f6')
    ax.set_title(f"Backtest equity ({os.path.basename(exp)}) — PnL {s['pnl_total']:+.2f} "
                 f"({s['pnl_pct']:+.2f}%), Sharpe {s['sharpe']:.2f}, {s['num_fills']} fills")
    ax.grid(alpha=.2); fig.autofmt_xdate()
    fig.savefig(os.path.join(FIGDIR,'fig2_backtest.png'), bbox_inches='tight', dpi=110)
    print('saved fig2_backtest.png ->', os.path.basename(exp))
else:
    print('no backtest export found under data/exports/')
```

## 4. Binary-option greeks: the implied risk-neutral distribution
Use the `analytics` module on a NatGas strike ladder (candlestick parquet) to recover the implied
settlement distribution (Breeden–Litzenberger) and the vol smile.

```python
gpath = os.path.join(ROOT,'analytics','out','greeks_demo.json')
if os.path.exists(gpath):
    g = json.load(open(gpath))
    snap = min((s for s in g['snapshots'] if s.get('strikes')),
               key=lambda s: (s.get('fit_rmse',9e9), -s.get('n_quoted',0)))
    ks = [k for k in snap['strikes'] if k.get('quoted')]
    K = [k['strike'] for k in ks]
    fig, (a1,a2) = plt.subplots(1,2,figsize=(12,4))
    a1.bar(K, [k.get('density') or 0 for k in ks], width=(K[1]-K[0])*0.8 if len(K)>1 else .01,
           color='#8b5cf6', alpha=.6)
    a1b = a1.twinx(); a1b.plot(K, [k.get('cdf') or 0 for k in ks], color='#3b82f6'); a1b.set_ylim(0,1)
    a1.set_title(f"Implied distribution {g['event']}  F~{snap.get('F_est')}"); a1.set_xlabel('strike $')
    iv = [(k['strike'],k['iv']) for k in ks if k.get('iv') is not None]
    if iv: a2.plot([k for k,_ in iv],[v for _,v in iv],'o-',color='#f59e0b',ms=3)
    a2.set_title('Implied-vol smile'); a2.set_xlabel('strike $'); a2.grid(alpha=.2)
    fig.savefig(os.path.join(FIGDIR,'fig3_greeks.png'), bbox_inches='tight', dpi=110)
    m = snap.get('moments',{})
    print(f"implied mean={m.get('mean')} std={m.get('std')} skew={m.get('skew')}; saved fig3_greeks.png")
else:
    print('run analytics/greeks.py first to create greeks_demo.json')
```
